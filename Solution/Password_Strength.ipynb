# GOAL:
 The goal of this project is to use Artificial Neural Network to build a predictive model that will be able to calculate the strength of password entered by user.

# 1. IMPORTING LIBRARIES

```python
# Import libraries and packages
import matplotlib as plt
%matplotlib inline
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , accuracy_score, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
import warnings
import itertools
import time

warnings.filterwarnings("ignore")
SEED = 42   # set random seed
```

```python
# display all columns of the dataframe
pd.options.display.max_columns = None

# display all rows of the dataframe
pd.options.display.max_rows = 50 # None

# returns an output value upto 6 decimals
pd.options.display.float_format = '{:.6f}'.format
```

# 2.LOADING THE DATA

```python
# load the data into a dataframe variable
data = pd.read_csv('passwords.csv')
```

```python
# check the top of the dataframe
data.head()
```

```python
# create a copy of the data before further operations
df = data.copy(deep=True)
```

# 3.EXPLORING THE DATA

```python
# check the info of the dataframe
df.info()
```

```python
# check the target column unique values
df['strength'].value_counts()
```

# 4.FEATURE ENGINEERING

```python
# changing the dtype of password column to string
df['password'] = df['password'].astype(str)
```

```python
# code to create a list of all possible characters in the password column
import string
string.ascii_letters
chars = string.printable
print(chars)
```

```python
# adding columns to df with occurrences of each character in a password
for char in chars:    
    df[char] = df['password'].apply(lambda x: x.count(char))
```

```python
# check out top of the new dataframe
df.head()
```

# 5. FEATURE SELECTION

```python
# check some statistics of the new dataframe
stats = df.describe().T
stats
```

We can see some of the characters have zero occurrences, hence there is no point in keeping them

```python
# create a list to drop the character columns which have no occurrences in the passwords
cols_to_drop = stats[stats['max']==0].index
cols_to_drop
```

```python
# dropping the columns
df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
# checking the stats of the df again
df.describe()
```

```python
df.head()
```

# 6.DATA PARTITION

```python
# we separate the target and features
X = df.drop(['strength','password'], axis=1)
y = df['strength']
```

```python
# partition the data into 80% training and 20% validation
(trainX, testX, trainY, testY) = train_test_split(X, y, test_size=0.3, random_state=SEED)
```

```python
# convert the labels from integers/categories to vectors 
trainY = to_categorical(trainY, num_classes=3)   # fit_transform = find all unique class labels + transform into one-hot encoded labels
testY = to_categorical(testY, num_classes=3)     # transform = perform the one-hot encoding (unique class labels already found)
```

# 7. DEFINING THE ARCHITECTURE OF ANN MODEL

```python
# define the 3072-1024-512-1 architecture using tensorflow
model = Sequential()   

# input layer has 92 neurons as there are 92 features in our train dataset
# first hidden layer has 2048 nodes
model.add(Dense(units= 2048, input_shape=(92,), kernel_initializer = 'uniform', activation="relu"))  

# # dropout for second layer 
# model.add(Dropout(0.4))

# second hidden layer has 1024 nodes
model.add(Dense(units=1024, kernel_initializer='uniform', activation="relu"))     

# third hidden layer has 512 nodes
model.add(Dense(units=512, kernel_initializer='uniform', activation="relu")) 

# fourth hidden layer has 256 nodes
model.add(Dense(units=256, kernel_initializer='uniform', activation="relu"))

 # output layer with number of possible class labels
model.add(Dense(units=3,kernel_initializer='uniform', activation="softmax"))           
```

# 8.COMPILING TENSORFLOW ANN MODEL

```python
# initialize our initial learning rate and # of epochs to train for
INIT_LR = 0.01
EPOCHS = 100
 
# compile the model using SGD as our optimizer and categorical cross-entropy loss
# (you'll want to use crossentropy for multi-class classification)
print("[INFO] compiling network network...")
opt = SGD(lr=INIT_LR)   # Stochastic Gradient Descent (SGD) optimizer
model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
```

```python
# checking the summary of the model
model.summary()
```

# 9. TRAINING MODEL USING TRAINING DATA

```python
# train the neural network on training data set
# batch_size (32) controls the size of each group of data to pass through the network. 

time1 = time.time()   # to measure time taken
H = model.fit(trainX, trainY, validation_data=(testX, testY), epochs=EPOCHS, batch_size=1000)
print('Time taken: {:.1f} seconds'.format(time.time() - time1))   # to measure time taken
```

# 10.EVALUATING MODEL ON TEST DATA

```python
# evaluate the network
print("[INFO] evaluating network...")
pred_prob = model.predict(testX, batch_size=100)
```

**Convert testY and y_pred into 0's, 1's and 2's for classification report**

```python
# Note: 0->Weak, 1->Strong, 2->Very Strong
```

```python
test_y = [ np.argmax(i)  for i in testY]
pred_y = [ np.argmax(i)  for i in pred_prob]
```

```python
def plot_confusion_metrix(y_true, y_pred,classes,
                         normalize=False,
                         title='Confusion Matrix',
                         cmap=plt.cm.Blues):
    """
    Objective
    ----------
    plot confussion matrix, classification report and accuracy score
    
    parameters
    ----------
    y_true : array-like of shape (n_samples,)
        Ground truth (correct) target values.

    y_pred : array-like of shape (n_samples,)
        Estimated targets as returned by a classifier.
    
    classes : list
        List of labels to index the matrix
        
    title : title for matrix
    cmap : colormap for matrix 
    
    returns 
    ----------
   all accruacy metrics 
    """
    
    
    cm = confusion_matrix(y_true,y_pred)
    
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized Confusion Matrix")
    else:
        print("Confusion Matrix, Without Normalisation")

    
    plt.imshow(cm, interpolation='nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks,classes,rotation=35)
    plt.yticks(tick_marks,classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() /2.
    
    for i , j in itertools.product(range(cm.shape[0]), range(cm.shape[0])):
        plt.text(j, i, format(cm[i,j], fmt),
                 horizontalalignment='center',
                 color='white' if cm[i, j] > thresh else 'black')
    
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    # plt.tight_layout()
    
    plt.show()
    
    print("-----------------------------------------------------")
    print('Classification report')
    print(classification_report(y_true,y_pred))
    
    print("-----------------------------------------------------")
    acc= accuracy_score(y_true,y_pred)
    print("Accuracy of the model: ", acc)
```

```python
# plot confusion matrix for the prediction results
plot_confusion_metrix(test_y, pred_y,classes=["0:Weak","1:Strong","2:Very Strong"])
```

**We can see that our model gives great accuracy, but we can see that number of strong cases out of the misclassified passwords is large, but since they are misclassified as weak it will only prompt user to choose more complex password, hence we can ignore these errors.**

Also we should not only depend on above metrics for evaluation of our model, we also need to check the way in which the model was trained in each epoch

# 11.PLOTTING ACCURACY AND VALIDATION LOSS

```python
# plot the training and validation loss
N = np.arange(0, EPOCHS)
plt.style.use("ggplot")
plt.figure(figsize = [10,8])
plt.plot(N, H.history["loss"], label="train_loss")
plt.plot(N, H.history["val_loss"], label="val_loss")
plt.title("ANN: Training & Validation Loss")
plt.xlabel("Epoch #", weight="bold")
plt.ylabel("Loss", weight="bold")
plt.legend()
plt.show()
```

```python
# plot the training and validation accuracy
N = np.arange(0, EPOCHS)
plt.style.use("ggplot")
plt.figure(figsize = [10,8])
plt.plot(N, H.history["accuracy"], label="train_acc")
plt.plot(N, H.history["val_accuracy"], label="val_acc")
plt.title("ANN: Training and Validation Accuracy")
plt.xlabel("Epoch #", weight="bold")
plt.ylabel("Accuracy", weight="bold")
plt.legend()
plt.show()
```

**We can see that there aren't any gaps in both training and validation accuracy and loss plots and both graphs are proceeding towards a point of stability, which indicates our model is trained well. Also, we need not train it after 40 epochs since there is no significant increase/decrease in accuracy and loss**

# 12.TRAINING THE FINAL MODEL FOR DEPLOYMENT

```python
# initialize our initial learning rate and # of epochs to train for
INIT_LR = 0.01
EPOCHS = 40
# compile the model using SGD as our optimizer and categorical cross-entropy loss
# (you'll want to use crossentropy for multi-class classification)
print("[INFO] compiling network network...")
opt = SGD(lr=INIT_LR)   # Stochastic Gradient Descent (SGD) optimizer
model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
# train the neural network on training data set
# batch_size (32) controls the size of each group of data to pass through the network. 

time1 = time.time()   # to measure time taken
H = model.fit(trainX, trainY, validation_data=(testX, testY), epochs=EPOCHS, batch_size=1000)
print('Time taken: {:.1f} seconds'.format(time.time() - time1))   # to measure time taken
```
